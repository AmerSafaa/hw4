```python
# Age Demographics

# The below each broken into bins of 4 years (i.e. &lt;10, 10-14, 15-19, etc.) 
# Create the bins in which Data will be held
bins = [0, 10, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the four bins
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+' ]

# Cut Age and place the players into bins
pd.cut(purchase["Age"], bins, labels=group_names)

purchase["Age Summary"] = pd.cut(purchase["Age"], bins, labels=group_names)

# Creating a group based off of the bins
purchase_groups = purchase.groupby("Age Summary")

#   * Totals
grouped_df = purchase.groupby(['Age Summary'])
ByAge_count = grouped_df["Age Summary"].count()

#  * Purchase Count
count_age = purchase['Age Summary'].count()

#   * Average Purchase Price
ByAge_price = grouped_df["Price"].mean()

#   * Total Purchase Value
ByAge_purchase = ByAge_count * ByAge_price

#   * Normalized Totals
ByAge_Total = ByAge_purchase/count_age



#Create Summary Table
state_summary_table2 = pd.DataFrame({"Age Summary":purchase_groups,
                                     "Purchase Count":ByAge_count,
                                     "Average Purchase Price":ByAge_price, 
                                     "Total Purchase Value":ByAge_purchase, 
                                     "Normalized Totals Revenue":ByAge_Total})
#state_summary_table2.head



```
